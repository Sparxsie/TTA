```python
# task 3
# print mean

import pandas as pd 

holData = pd.read_csv("HolidayRatings.csv")

print ("Mean number of all inclusive hotels in all destinations is", int(holData["All Inc"].mean()))
```

```python

# task 4
# print highest scoring destination
import pandas as pd 
holData = pd.read_csv("HolidayRatings.csv")


print ("The highest scoring destination entry is :\n", holData[holData["Rating"] == holData["Rating"].max()]["Dest"])
```

```python
# task 5
# find lowest scoring destination
import pandas as pd 
holData = pd.read_csv("HolidayRatings.csv")

print ("The lowest scoring destination entry is :\n", holData[holData["Rating"] == holData["Rating"].min()]["Dest"])
```
